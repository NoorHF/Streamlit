```python
import numpy as np
```

```python
from scipy.stats import zscore
```

```python
numeric_df = df.select_dtypes(include='number')
```

```python
numeric_df = numeric_df[(np.abs(zscore(numeric_df))< 3).all(axis = 1)]
```

```python
numeric_df.shape
```

```python
categorical_df = df.select_dtypes(include = 'object')
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
le = LabelEncoder()
categorical_columns = ['REGION','TENURE', 'MRG','TOP_PACK' ]
categorical_df[categorical_columns] = categorical_df[categorical_columns].apply(lambda col: le.fit_transform(col))
```

```python
joined_df = categorical_df.join(numeric_df, how = 'inner')
```

```python
joined_df.shape
```

```python
joined_df
```

```python
X = joined_df.drop(columns = ['CHURN','user_id'])
Y = joined_df['CHURN']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split ( X, Y, test_size=0.2, random_state= 70)
```

```python
model = RandomForestClassifier(n_estimators= 5, max_depth=5)
```

```python
model.fit(X_train, Y_train)
```

```python
prediction = model.predict(X_test)
```

```python
accuracy_score(Y_test, prediction)
```
